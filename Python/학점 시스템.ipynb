```python
import sys

def grade(mid, fin):
    avg=(int(mid)+int(fin))/2

    if avg >= 90:
        g= 'A'
    elif avg>= 80:
        g= 'B'
    elif avg >= 70:
        g= 'C'
    elif avg >= 60:
        g='D'
    else:
        g='F'
    return str(avg), str(g)
```

```python
def show(grade='base'):
    sort_s1=sorted(students.items(), key = lambda a: a[1][3], reverse=True)
    a = ['Student', 'Name', 'Midterm', 'Final', 'Average', 'Grade']
    print('%10s %13s %10s %8s %10s %8s' % (a[0], a[1], a[2], a[3], a[4], a[5]), '\n')
    print('----------------------------------------------------------------------------\n\n')
    for i in sort_s1:
        if grade == 'base' or i[1][4] == grade :
            print('%10s %13s %10s %8s %10s %8s' % (i[0], i[1][0], i[1][1], i[1][2], i[1][3], i[1][4]), '\n')
```

```python
def add():
    num = int(input('Student ID : '))
    if num in students.keys():
        print('ALREADY EXISTS')
    else:
        name=(input('Name :'))
        mid=(input('Midterm Score : '))
        fin=(input('Final Score : '))
        print('Student added.')
        avg,g=grade(mid,fin)
        students[num]=[name,mid,fin,avg,g]
```

```python
def remove():
    if len(students.values()) == 0:
        print('List is empty')
    else:
        num=int(input('Student Id : '))
        if num not in students.keys():
            print('NO SUCH PERSON')
        else :
            del students[num]
            print('Student removed')
```
