```python
import numpy as np
import scipy.stats as sp
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest
from scipy import stats
```

```python
df=pd.DataFrame({'sample':[74.5,81.2,73.8,82,76.3,75.7,80.2,72.6,77.9,82.8]})
```

```python
tt=stats.ttest_1samp(df,76.7)
t,p=tt.statistic.round(3),tt.pvalue.round(3)
print('t검정통계량' , t)
print('p_value', p)
```

```python
df=pd.DataFrame({'hsg':[270,260,236,234],'ss':[228,285,225,262],'spa':[277,284,231,208]})
chi,pval,dof,expected=stats.chi2_contingency(df.T)
print('chisq',chi)
print('df',dof)
print('p',pval)
print('expected value \n',expected)
```

```python
left=pd.DataFrame({'key1':['k0','k1','k2','k3'],'key2':['k0','k1','k0','k1'],'a':['a0','a1','a2','a3'],\
                  'b':['b0','b1','b2','b3']})
right=pd.DataFrame({'key1':['k0','k1','k2','k3'],'key2':['k0','k0','k0','k0'],'c':['c0','c1','c2','c3'],\
                  'd':['d0','d1','d2','d3']})
```

```python
result=pd.merge(left,right,on=['key1','key2'],how='left')
result
```

```python
result=pd.merge(left,right,on=['key1','key2'],how='right')
result
```

```python
result=pd.merge(left,right,on=['key1','key2'],how='left',indicator=True)
result
```
