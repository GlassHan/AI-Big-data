```python
df['Sex']=pd.get_dummies(df['Sex'])
```

```python
name=pd.get_dummies(df['Name'])
ticket=pd.get_dummies(df['Ticket'])
cabin=pd.get_dummies(df['Cabin'])
embarked=pd.get_dummies(df['Embarked'])
```

```python
df=df.drop(['Name','Ticket','Cabin','Embarked'],axis=1)
```

```python
df=pd.merge(df,name,how='outer',left_index=True,right_index=True)
df=pd.merge(df,ticket,how='outer',left_index=True,right_index=True)
df=pd.merge(df,cabin,how='outer',left_index=True,right_index=True)
df=pd.merge(df,embarked,how='outer',left_index=True,right_index=True)
```

```python
df=df.dropna()
```
