# 다중 회귀분석 실습

```python
df=pd.read_csv('DIABETEST.csv',engine='python',encoding='949')
df.head()
```

```python
sns.pairplot(df)
#그래프를 통한 다중공선성 확인. CHOLESTEROL 와 LDL 사이에 양의 상관 관계가 있음을 볼 수 있다.
#또한 HDL와 TCH 사이에 음의 상관관계가 보인다.
```

```python
sns.pairplot(df,y_vars='Y',x_vars=['BMI','TCH','LTG','GLUCOSE'])
```

```python
df.corr().round(3)
#그래프와 마찬가지로 CHOLESTEROL 와 LDL 사이에 양의 상관 관계가 있음을 볼 수 있다.
#또한 HDL와 TCH 사이에 음의 상관관계가 보인다. LDL 과 TCH 사이에 약한 양의 상관관계가 보인다.
```

```python
model=smf.ols(formula='Y~AGE+GENDER+BMI+BP+CHOLESTEROL+LDL+HDL+TCH+LTG+GLUCOSE',data=df)
result=model.fit()
print(result.summary())

#전체 변수를 넣고 상관분석을 한 결과 회귀식의 p-value가 0.05보다 작으므로 유의하고
#결정계수 값이 0.518로 설명력이 낮다.
```

```python
df_x=df.drop('Y',axis=1)
df_x_const=add_constant(df_x)
df_vif=pd.DataFrame()
df_vif['variable']=df_x_const.columns
df_vif['VIF']=[variance_inflation_factor(df_x_const.values,i)for i in range(df_x_const.shape[1])]
df_vif.sort_values('VIF',inplace=True)
df_vif.round(3)

#변수들 간의 다중공선성을 분석한 결과 LTG ,HDL,LDL ,CHOLESTEROL 가 다중공선성이 있다는 것을 알 수 있다.
```

```python
model=smf.ols(formula='Y~AGE+GENDER+BMI+BP+TCH+GLUCOSE',data=df)
result=model.fit()
print(result.summary())

#다중공선성을 가진  LTG ,HDL,LDL ,CHOLESTEROL 변수를 제외하고 회귀식을 다시 만들었다.
#그 결과 p-value가 0.05보다 작으므로 다음 회귀식은 유의하다.
#전에 만든 회귀식과 비교해서 결정계수가 0.449로 작아졌다. 즉 모델의 설명력이 약간 감소되었다.
```

```python
df_x=df.drop(['Y','CHOLESTEROL','LDL','HDL','LTG'],axis=1)
df_x_const=add_constant(df_x)
df_vif=pd.DataFrame()
df_vif['variable']=df_x_const.columns
df_vif['VIF']=[variance_inflation_factor(df_x_const.values,i)for i in range(df_x_const.shape[1])]
df_vif.sort_values('VIF',inplace=True)
df_vif.round(3)

#남은 변수들 간의 다중공선성을 분석한 결과 나머지 변수들은 다중공선성이 낮은걸로 판단된다.
```

```python
df_y=df['Y']
model=LinearRegression()
rfe=RFE(estimator=model,n_features_to_select=5).fit(df_x,df_y)
selected_cols=df_x.columns[rfe.support_]
removed_cols=df_x.columns[~rfe.support_]
print('selected variable : {}'.format(selected_cols))
print('removed variable : {}'.format(removed_cols))

#후진제거법으로 변수5개'GENDER', 'BMI', 'BP', 'TCH', 'GLUCOSE'를 선택하였다. 
```

```python
model=smf.ols(formula='Y~GENDER+BMI+BP+TCH+GLUCOSE',data=df)
result=model.fit()
print(result.summary())

#후진제거법을 통해 선택한 변수로 회귀식을 만들었다.
#이때 p-value 값이 0.05 보다 작으므로 회귀식은 유의하다.
#또한 결정계수 값은 변하지 않았으며 adj결정계수는 0.001정도 증가 하였다.
```

```python
df_x_rfe=df_x[selected_cols]
df_x_rfe_const=add_constant(df_x_rfe)
df_rfe_vif=pd.DataFrame()
df_rfe_vif['variable']=df_x_rfe_const.columns
df_rfe_vif['VIF']=[variance_inflation_factor(df_x_rfe_const.values,i)for i in range(df_x_rfe_const.shape[1])]
df_rfe_vif.sort_values('VIF',inplace=True)
df_rfe_vif.round(3)

#후진제거법을 통해 선택한 변수 간의 다중공선성이 낮게 나온것을 확인 할 수 있다.
```
